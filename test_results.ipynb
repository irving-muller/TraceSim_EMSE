```python
from scipy.stats import wilcoxon, brunnermunzel
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
import numpy as np
from IPython.display import display, HTML
from matplotlib.ticker import FormatStrFormatter





def rakah_evaluation(results, metric_name, alternatives=[], dataset_name="", alternative="two-sided", best_method="TraceSim", y_lim=None, y_lim_diff=None):
    sns.set(rc={'figure.figsize':(16.8,4.1)})
    sns.set_theme(style="whitegrid")
    f, axes = plt.subplots(1, 2)

    # Transform to dataframe
    data = []
    
    
    for i, res in enumerate(results):
        data.extend([{'exp': res['name'], 'e':0, metric_name.upper(): r, 'chunk_id': int(chunk_id)} for chunk_id, r in res[metric_name].items()])
    
    
    # Perform the statistical tests between original and the other ones
    panda_results = pd.DataFrame(data)
    wilcoxon_results = {}
    original = None
    
    for i, res in enumerate(results):
        if res['name'] == best_method:
            original = res
    
    
    diff_results = {}
    
    for i, res in enumerate(results):
        if res['name'] == best_method:
            continue
            
        n_chunks = (panda_results['chunk_id'].max() + 1)
        x1 = []
        x2 = []
        ignored_chunks = []
        
        for chunk_id in range(n_chunks):
            f1 = original[metric_name].get(str(chunk_id))
            f2 = res[metric_name].get(str(chunk_id))
            
            
            if f1 is None or f2 is None:
                ignored_chunks.append(chunk_id)
                continue
                
            x1.append(f1)
            x2.append(f2)
            
        if len(ignored_chunks) > 0:
            print(f"WARNING: chunks are missing in {original['name']} vs{res['name']}: {ignored_chunks}")
        wilcoxon_results[res['name']] = wilcoxon(x1,x2,alternative=alternative, correction=True, zero_method="zsplit")
        diff_results[res['name']] = [a-b for a,b in zip(x1,x2)]
        
        
    sns.set_theme(style="whitegrid")
    
    ax = sns.violinplot(x="exp", y=metric_name.upper(), data=panda_results, inner="quart", linewidth=1.5, palette=sns.color_palette("pastel"), showmeans=True, 
                         meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"white",
                          "markersize":"4"},ax=axes[0])
    

#     ax = sns.boxplot(x="exp", y=metric_name.upper(), data=panda_results)
#     ax = sns.scatterplot(x="exp", y=metric_name.upper(), data=means, edgecolor='white',size=5.0, linewidth=2.0)    
    xticks_labels = []
    xticks_x_positions = []
    
    for txt in ax.get_xticklabels():
        wr =  wilcoxon_results.get(txt.get_text())
        
        iden = ""

            
        txt.set_text(txt.get_text() + iden)
        xticks_x_positions.append(txt.get_position()[0])
        xticks_labels.append(txt)
        
    ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
    ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
        
    means = panda_results.groupby(['exp'], as_index=False).mean()
    
    y_points = [ means[means.exp==lab.get_text()].iloc[0][metric_name.upper()] for lab in xticks_labels]    
    ax.plot(xticks_x_positions,y_points,'wo', markersize=4)     
        
    
    
    ax.set_xticklabels(xticks_labels, {'fontsize':15})
    ax.set_xlabel("")
    ax.set_ylim(y_lim)
    ax.set_ylabel(metric_name.upper(), {'fontsize':11.5})
    ax.tick_params(axis='y', which='major', labelsize=13)
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    
 
    table = []
    means = panda_results.groupby(['exp'], as_index=False).mean()
    
    for i, res in enumerate(results):
        name = res['name']

        
        if name == best_method:
            diff = np.asarray([0.0])
            wil_p_value = None
        else:
            diff = np.asarray(diff_results[name])
            wil_p_value = wilcoxon_results[res['name']].pvalue
            
            
        table.append({"Method": res['name'],
                 f"Mean. {metric_name}": means[means.exp==res['name']].iloc[0][metric_name.upper()],     
                 f"Mean Diff. {metric_name}": diff.mean() * 100,
                 "Wilcoxon p-value": wil_p_value if wil_p_value else "--",
                 })
            
    
            
    table = pd.DataFrame(table)
    
    display(HTML(table.to_html(index=False)))
    
    
    
    table = []
    
    
    for exp_name, (z,p_value) in wilcoxon_results.items():
        if best_method == exp_name:
            continue
        for diff in  diff_results[exp_name]:
            
            table.append({
                "exp": exp_name,
                "metric": metric_name,
                "dt_name": dataset_name,
                "p_value": p_value,
                "diff": diff
            })

    table = pd.DataFrame(table)
    
    ax = sns.boxplot(x="exp", y="diff", data=table, fliersize=1.5, linewidth=2.0,showmeans=True, 
                         meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"white",
                          "markersize":"4"}, palette=sns.color_palette("pastel")[1:],ax=axes[1])
    
    xticks_labels = []
    for txt in ax.get_xticklabels():
        p_value = wilcoxon_results[txt.get_text()][1]
        iden = ""
        
        
        if p_value is not None and p_value < 0.01:
            iden += "\\bigstar"
            
        if len(iden) > 0:
            iden = "$^{" + iden +"}$"
        else:
            iden = "$^{\\,}$"
            
        
        txt.set_text(txt.get_text() + iden)
        xticks_labels.append(txt)
        
    
    ax.set_xticklabels(xticks_labels, {'fontsize':15})
    
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

    ax.set_xlabel("")
    ax.set_ylabel("Δ\u2009" + metric_name.upper(), {'fontsize':11.5},labelpad=0.0)
    ax.set_ylim(y_lim_diff)
    
    if y_lim_diff is not None:
        start = 0
        
        for i in np.arange(-1, 1, 0.05):
            if i < y_lim_diff[0]:
                continue
            start = i
            break
            
        
        ax.set_yticks(np.arange(start, y_lim_diff[1]+0.0001, 0.05))

    ax.tick_params(axis='y', which='major', labelsize=13)
    
    
    f.tight_layout()
    
    plt.savefig(f"plots/{dataset_name}_{metric_name}_with_diff.pdf", format='pdf', bbox_inches='tight', pad_inches=0)
    
    
    
    

    return wilcoxon_results, diff_results
```
