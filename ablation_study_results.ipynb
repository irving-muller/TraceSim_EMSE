```python
from scipy.stats import wilcoxon, brunnermunzel
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
import numpy as np
from IPython.display import display, HTML





def rakah_evaluation(results, metric_name, original_idx=0, alternatives=[], dataset_name="", alternative="two-sided"):
    fig = plt.figure()

    sns.set(rc={'figure.figsize':(16,8)})
    
    
    # Transform to dataframe
    data = []
    
    
    for i, res in enumerate(results):
#         panda_results = [{'exp': original['name'], 'e':0, metric_name: r, 'chunk_id': int(chunk_id)} for chunk_id, r in original[metric_name].items()]
        data.extend([{'exp': res['name'], 'e':0, metric_name.upper(): r, 'chunk_id': int(chunk_id)} for chunk_id, r in res[metric_name].items()])
    
    
    # Perform the statistical tests between original and the other ones
    panda_results = pd.DataFrame(data)
    wilcoxon_results = {}
    wilcoxon_results_less = {}
    original = results[original_idx]
    diff_results = {}
    
    for i, res in enumerate(results):
        if i == original_idx:
            continue
            
        n_chunks = (panda_results['chunk_id'].max() + 1)
        x1 = []
        x2 = []
        ignored_chunks = []
        
        for chunk_id in range(n_chunks):
            f1 = original[metric_name].get(str(chunk_id))
            f2 = res[metric_name].get(str(chunk_id))
            
            
            if f1 is None or f2 is None:
                ignored_chunks.append(chunk_id)
                continue
                
            x1.append(f1)
            x2.append(f2)
            
        if len(ignored_chunks) > 0:
            print(f"WARNING: chunks are missing in {original['name']} vs{res['name']}: {ignored_chunks}")
                                                                                                            
        wilcoxon_results[res['name']] = wilcoxon(x1,x2,alternative=alternative, correction=True, zero_method="zsplit")
        p_value =  wilcoxon_results[res['name']][1]

            
        
        diff_results[res['name']] = [a-b for a,b in zip(x1,x2)]
        
    
        
#     sns.set_theme(style="whitegrid")
    
#     ax = sns.violinplot(x="exp", y=metric_name.upper(), data=panda_results, inner="quart", linewidth=1.8)
#     sns.stripplot(x="exp", y=metric_name.upper(), data=panda_results, edgecolor='white', linewidth=0.5)

    
#     xticks_labels = []
#     for txt in ax.get_xticklabels():
#         wr =  wilcoxon_results.get(txt.get_text())

#         iden = ""
#         txt.set_text(txt.get_text() + iden)
        
#         xticks_labels.append(txt)
    
#     ax.set_xticklabels(xticks_labels, {'fontsize':15})
#     plt.xlabel("")
#     plt.ylabel(metric_name.upper(), fontsize=15)
    
#     plt.savefig(f"plots/ablation{dataset_name}_{metric_name}.pdf", format='pdf', bbox_inches='tight', pad_inches=0)
    
    
    table = []
    diff_table = []
    
    for i, res in enumerate(results):
        name = res['name']
        
        if i == original_idx:
            diff = np.asarray([0.0])
            wil_p_value = None
            bru_p_value = None
            cliff_delta = None
        else:
            diff = np.asarray(diff_results[name])
            wil_p_value = wilcoxon_results[res['name']].pvalue
            
        sorted_diff = sorted(diff)
        previous =  diff[0]
        
            

        table.append({"Method": res['name'],
                 f"Mean Diff. {metric_name}": diff.mean() * 100,
                 f"Median Diff. {metric_name}": np.median(diff * 100),
                 "Wilcoxon p-value": wil_p_value if wil_p_value else "--",
                 })
        
        if wil_p_value is not None:
            for d in diff:
                diff_table.append({
                    "exp": res['name'],
                     metric_name.upper(): d
                })
        
                
    table = pd.DataFrame(table)
    table.style.set_caption("{metric_name}")
    display(HTML(table.to_html(index=False)))
    
    
    diff_table = pd.DataFrame(diff_table)
    
    
    
    return wilcoxon_results, diff_results
    
```

```python
# Unroll
table = []

for dt_name, res_by_metric in diff_wilcoxon_results.items():
    for metric_name, (wilcoxon_value,diff_results) in res_by_metric.items():
        for exp_name, (z,p_value) in wilcoxon_value.items():
            for diff in  diff_results[exp_name]:
                table.append({
                    "exp": exp_name,
                    "dt_name": dt_name,
                    "metric": metric_name,
                    "dt_name": dt_name,
                    "p_value": p_value,
                    "diff": diff
                })
                
table = pd.DataFrame(table)
```

```python
def plot_diff_cmp(table, exp_name, metric_name, diff_wilcoxon_results, short_name=None, axes=None, y_lim=None):
    import re
    panda_results = table[(table.exp == exp_name) & (table.metric == metric_name)]
    
    
    sns.set_theme(style="whitegrid")
    ax = sns.boxplot(x="dt_name", y="diff", data=panda_results, fliersize=1.5, linewidth=2.0, ax=axes,showmeans=True, 
                         meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"white",
                          "markersize":"4"}, palette=sns.color_palette("pastel"))

    xticks_labels = []
    mean_diff = []
    for txt in ax.get_xticklabels():
        wilcoxon_v, diff_results = diff_wilcoxon_results[txt.get_text()][metric_name]
        p_value = wilcoxon_v[exp_name][1]
        
        mean_diff.append((txt.get_text(), round(np.array(diff_results[exp_name]).mean(), 4), p_value))
        
        
        iden = ""
        
        
        if p_value is not None and p_value < 0.01:
            iden += "\\bigstar"
            
        if len(iden) > 0:
            iden = " $^{" + iden + "}$"
        else:
            iden = " $^{\\,}$"
            
            
        txt.set_text(txt.get_text() + iden)
        xticks_labels.append(txt)
        
    
        
    ax.set_xticklabels(xticks_labels)
    
    if axes is None:
        axes = plt

    axes.set_xlabel("")
    axes.set_ylabel("$\Delta$ " + metric_name.upper())
    axes.set_ylim(y_lim)
    
    norm_exp_name = re.sub(r"[ $\\/]", "", exp_name)
    
    if short_name is None:
        short_name = norm_exp_name

    


```

```python
plt.rc('text', usetex=False)

f, axes = plt.subplots(1, 3)
sns.set(rc={'figure.figsize':(18,6)})

plot_diff_cmp(table, "(1) w/o $\operatorname{gw}(\cdot)$", "AUC",diff_wilcoxon_results, "gw", axes=axes[0], y_lim=[-0.12,0.23])
plot_diff_cmp(table, "(1) w/o $\operatorname{gw}(\cdot)$", "MAP",diff_wilcoxon_results, "gw", axes=axes[1], y_lim=[-0.12,0.23])
plot_diff_cmp(table, "(1) w/o $\operatorname{gw}(\cdot)$", "RR@1",diff_wilcoxon_results, "gw", axes=axes[2], y_lim=[-0.12,0.23])

f.tight_layout() 
plt.savefig(f"plots/ablation_diff_gw.pdf", format='pdf', bbox_inches='tight', pad_inches=0)
```
